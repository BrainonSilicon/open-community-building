# Generate wordcoud for CM team's day-to-day work

This is a wordclod generator for a ';' separated table with keywords and frequency for text size in the wordcloud.

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
# Importing Dataset
df = pd.read_csv("cm-words.csv", sep=";")
df
```

```python
#Creating the text variable
tuples = [tuple(x) for x in df.values]

#We can generate the wordcloud in the following manner.
# Generate word cloud
word_cloud = WordCloud(
    width=3000,
    height=2000,
    random_state=1,
    background_color="white",
    collocations=False,
    stopwords=STOPWORDS,
    ).generate_from_frequencies(dict(tuples))


# Display the generated Word Cloud
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```
